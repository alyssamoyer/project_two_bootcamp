```python
import pandas as pd
import json 
from datetime import date
from datetime import datetime
import calendar
```

```python
#read csv into pandas
df_M6 = pd.read_csv("fire_archive_M6_96619.csv",encoding = 'ISO-8859-1')  
df_M6
```

```python
df_M6['acq_date'] =  pd.to_datetime(df_M6['acq_date'], format='%Y-%m-%d')
df_M6
```

```python
#Create bins for utc times, bin names are the category names, bin y_values will be used for the y value when charting
utc_time_bins = [0, 400, 800, 1200, 1600, 2000, 2400]
bin_names = ["0 - 4:00 UTC", "4:00 - 8:00 UTC", "8:00 - 12:00 UTC", "12:00 - 16:00 UTC", "16:00 - 20:00 UTC", "20:00 - 24:00 UTC"]
#Create new column that stores the time bins
df_M6["time_ranges"] = pd.cut(df_M6["acq_time"], utc_time_bins, labels=bin_names)

df_M6
```

```python
index_count = 0
df_M6['day_of_week'] = ""
for days in df_M6.iterrows():
    day = df_M6.loc[index_count, 'acq_date']
    day_name = calendar.day_name[day.weekday()]
    df_M6.loc[index_count, 'day_of_week'] = day_name
    day_num = day.weekday()
    index_count = index_count + 1
    

df_M6
```

```python
df_M6.to_csv('fire_archive_M6_96619_clean.csv')
```
