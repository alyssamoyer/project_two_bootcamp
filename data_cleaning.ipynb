```python
import pandas as pd
import json 
from datetime import date
from datetime import datetime
import calendar
```

```python
#read csv into pandas
df_M6 = pd.read_csv("fire_archive_M6_96619.csv",encoding = 'ISO-8859-1')  
df_M6
```

```python
df_max = df_M6.groupby(['acq_date'])['frp'].max()
df_max.to_csv('fire_archive_M6_96619_max.csv', header ="true")
```

```python
#convert the datatype of acq_date to date time
df_M6['acq_date'] =  pd.to_datetime(df_M6['acq_date'], format='%Y-%m-%d')
df_M6
```

```python
#Create bins for utc times, bin names are the category names, bin y_values will be used for the y value when charting
utc_time_bins = [0, 400, 800, 1200, 1600, 2000, 2400]
bin_names = ["0 - 4:00 UTC", "4:00 - 8:00 UTC", "8:00 - 12:00 UTC", "12:00 - 16:00 UTC", "16:00 - 20:00 UTC", "20:00 - 24:00 UTC"]
#Create new column that stores the time bins
df_M6["time_ranges"] = pd.cut(df_M6["acq_time"], utc_time_bins, labels=bin_names)

df_M6
```

```python
#loop through dataframe and add a new column for day of week. Use the calemdar to calculate the weekday from acq_date
index_count = 0
df_M6['day_of_week'] = ""
for days in df_M6.iterrows():
    day = df_M6.loc[index_count, 'acq_date']
    day_name = calendar.day_name[day.weekday()]
    df_M6.loc[index_count, 'day_of_week'] = day_name
    day_num = day.weekday()
    index_count = index_count + 1
    

df_M6
```

```python
#group the dataframe by day of week and time bins
df_M6_grouped = df_M6.groupby(['day_of_week', 'time_ranges'])['frp'].mean().reset_index()
df_M6_grouped.head()
```

```python
df_M6 =df_M6.drop(columns=['time_ranges'])
```

```python
#export csv of clean dataframe
df_M6.to_csv('fire_archive_M6_96619_clean.csv')
```

```python
#group the dataframe by day of week and time bins
df_M6_grouped = df_M6.groupby(['day_of_week', 'time_ranges'])['frp'].mean().reset_index()

df_M6_grouped.head()
```

```python
#loop through dataframe and add a new column for the x values based on the day of week. The x values will be used in high charts.
index_count = 0;
df_M6_grouped['x_value'] = ""
for rows in df_M6_grouped.iterrows():
    weekday = df_M6_grouped.loc[index_count, 'day_of_week']
    if weekday == "Monday":
        df_M6_grouped.loc[index_count, 'x_value'] = 0
    elif weekday == "Tuesday":
        df_M6_grouped.loc[index_count, 'x_value'] = 1
    elif weekday == "Wednesday":
        df_M6_grouped.loc[index_count, 'x_value'] = 2
    elif weekday == "Thursday":
        df_M6_grouped.loc[index_count, 'x_value'] = 3
    elif weekday == "Friday":
        df_M6_grouped.loc[index_count, 'x_value'] = 4
    elif weekday == "Saturday":
        df_M6_grouped.loc[index_count, 'x_value'] = 5
    elif weekday == "Sunday":
        df_M6_grouped.loc[index_count, 'x_value'] = 6
    index_count = index_count + 1

df_M6_grouped.head()
```

```python
#loop through the dataframe and add a new column for the y value based on time bins. The y value will be used in high charts.
index_count = 0;
df_M6_grouped['y_value'] = ""
for rows in df_M6_grouped.iterrows():
    time = df_M6_grouped.loc[index_count, 'time_ranges']
    if time == "0 - 4:00 UTC":
        df_M6_grouped.loc[index_count, 'y_value'] = 0
    elif time == "4:00 - 8:00 UTC":
        df_M6_grouped.loc[index_count, 'y_value'] = 1
    elif time == "8:00 - 12:00 UTC":
        df_M6_grouped.loc[index_count, 'y_value'] = 2
    elif time == "12:00 - 16:00 UTC":
        df_M6_grouped.loc[index_count, 'y_value'] = 3
    elif time == "16:00 - 20:00 UTC":
        df_M6_grouped.loc[index_count, 'y_value'] = 4
    elif time == "20:00 - 24:00 UTC":
        df_M6_grouped.loc[index_count, 'y_value'] = 5
    index_count = index_count + 1

df_M6_grouped ['frp'] = df_M6_grouped ['frp'] .fillna(value=0)
df_M6_grouped.head()
```

```python
#Add a new column that concatinates the x, y , and frp value. This will be spilt in javascript.
df_M6_grouped["hc_values"] = df_M6_grouped ["x_value"].astype(str) + "," + df_M6_grouped ["y_value"].astype(str) + "," + df_M6_grouped ["frp"].astype(str)

df_M6_grouped
```

```python
df_M6_grouped =df_M6_grouped.drop(columns=['time_ranges'])
```
